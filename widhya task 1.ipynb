```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```
# Reading the data from the provided link
url = "https://raw.githubusercontent.com/WidhyaOrg/datasets/master/bitcoin_dataset.csv"
df = pd.read_csv(url)
print("Data imported successfully")
df.head()
```

```
# Microtask - 4
df.isna().any() # To check which column has NaN values.

# True indicates that those columns have NaN values
# we will replace the NaN values with mean of the column values

df = df.fillna(df.mean())
```

```
# Microtask - 5
x = df.drop(['btc_market_price','Date'], axis=1) # taking all the feature column except date and btc_market_price
y = df[['btc_market_price']] # output response
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3) # spilting dataset into training and testing with 70% and 30%
linear_reg = LinearRegression() # linear regression instance
linear_reg.fit(x_train,y_train) # training the model
y_predict = linear_reg.predict(x_test) # testing the model on test data
```

```
# Calculating the mean squared error
mean_squared_error(y_test,y_predict)
```
